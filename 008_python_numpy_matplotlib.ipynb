# Дополнительные красивости для Matplotlib

```python
x = np.linspace(-5,5,1000)

plt.plot(x, x)
plt.plot(x, x**2)
plt.plot(x, x**3)
```

```python
x = np.linspace(-5,5,1000)

plt.plot(x, x, label="y=x")
plt.plot(x, x**2, label="y=x^2")
plt.plot(x, x**3, label="y=x^3")
plt.legend()
```

```python
plt.figure(figsize=(3,3), dpi=150)

x = np.linspace(-5,5,1000)

plt.plot(x, x, label="$y=x$") # Это позволяет задавать LaTeX формулы
plt.plot(x, x**2, label="$y=x^2$")
plt.plot(x, x**3, label="$y=x^3$")
plt.legend()

plt.title("Графики функций")
plt.xlabel("X-axis")
plt.ylabel("Y-axis");
```
